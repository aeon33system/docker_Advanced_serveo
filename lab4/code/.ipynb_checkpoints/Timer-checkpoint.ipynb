```python
from linebot import (
    LineBotApi
)
from linebot.exceptions import (
    InvalidSignatureError
)
from linebot.models import (
    TextSendMessage
)
import datetime
import pymysql
import json
import threading
from threading import Thread
import time

#--使用json讀取secretFile.txt
secretFile = json.load(open("./secretFile.txt",'r'))
#--從secretFile.txt取得channelAccessToken
channelAccessToken = secretFile['channelAccessToken']
#--從secretFile.txt取得channelAccessToken
line_bot_api = LineBotApi(channelAccessToken)
#--設定SQL位置等資料
config = {
          'host':secretFile['host'],
          'port':secretFile['port'],
          'user':secretFile['user'],
          'password':secretFile['passwd'],
          'db':secretFile['db'],
          }

#--創建一個連線資料庫的class，並啟動多執行序
class dbcnn(Thread):
    #--宣告一個連線資料庫事件
    def fun_db_open(self):
        #--繼承db,cursor 變數
        global db,cursor
        #--使用pymysql.connect 連線資料庫
        db = pymysql.connect(**config)
        #--建立遊標 ，cursor是處理數據的一種方法，為了查看或者處理結果集中的數據，游標提供了在結果集中一次一行或者多行前進或向後瀏覽數據的能力。可以把游標當作一個指針，它可以指定結果中的任何位置，然後允許用戶對指定位置的數據進行處理
        #--通俗來說就是，操作數據和獲取數據庫結果都要通過游標來操作。
        cursor = db.cursor();
    
    #--宣告一個中斷連線資料庫事件
    def fun_db_close(self):
        #--繼承db,cursor 變數
        global db,cursor
        #--關閉遊標
        cursor.close(); 
        #--關閉連線
        db.close(); 

#--將class實體化
dbcnn=dbcnn()

#--創建一個timer的class，並啟動多執行序
class timer(Thread):
    #--宣告一個檢查使用者的寵物飽食度低於X值時，透過LINE API傳遞資訊給使用者
    def fun_push(self):
        #--宣告timer_push每X秒執行一次fun_push事件
        timer_push = threading.Timer(3, self.fun_push)
        #--啟用SQL連線
        dbcnn.fun_db_open();
        SQL = "select Line_UID,hungry"
        SQL = SQL + ",(select Push_YN from ap_userdata s where m.Line_UID=s.Line_UID) as push"
        SQL = SQL + " from ap_userpet m where m.Del_YN='N' and m.hungry between '1' and '30' and m.status='1'"
        SQL = SQL + " and (select Push_YN from ap_userdata s where m.Line_UID=s.Line_UID)='Y'"
        #--將SQL指令傳遞給SQL並執行
        cursor.execute(SQL)
        #--取回SQL回傳的值
        data = cursor.fetchall(); 
        #--結束SQL連線
        dbcnn.fun_db_close();
        if len(data) > 0:
            #print("寵物還活著")
            #--將訊息傳遞給使用者
            for x in range(0,len(data)):
                data2=data[x]
                print(str(data2[0]) + "您可愛的寵物飽食度已剩餘" + str(data2[1]) + "%，請盡速餵食。")
                #ine_bot_api.push_message(data[x], TextSendMessage(text='寵物還活著'))
            #line_bot_api.push_message(data[0], TextSendMessage(text='寵物還活著'))
        timer_push.start()
        
     #--宣告一個每X秒扣除飽食度的事件
    def fun_hungry(self):
        #--宣告timer_hg每X秒執行一次fun_hungry事件
        timer_hg = threading.Timer(3, self.fun_hungry)
        #--啟用SQL連線
        dbcnn.fun_db_open();
        #--將SQL指令傳遞給SQL並執行
        cursor.execute("select ID,hungry from ap_userpet where Del_YN='N' and hungry > 0 and status='1'")
        #--取回SQL回傳的值
        data = cursor.fetchone(); 
        #--結束SQL連線
        dbcnn.fun_db_close();
        print(data)
        #--判斷若有從SQL取得值
        if data is not None:
            #--啟用SQL連線
            dbcnn.fun_db_open();
            SQL = "UPDATE ap_userpet SET hungry = hungry - 10 WHERE Del_YN='N' and hungry > 0 and status='1'"
            #--將SQL指令傳遞給SQL並執行
            cursor.execute(SQL);
            #--SQL執行後確實動作(select無須commit，但UPDATE、INSERT INTO若不使用commit則不會存儲到資料庫)
            db.commit();
            #--結束SQL連線
            dbcnn.fun_db_close();print("-1")
        #--判斷如果有寵物飽食度為0，且狀態為正常，將狀態改為逃走
        dbcnn.fun_db_open();
        cursor.execute("select ID,hungry from ap_userpet where Del_YN='N' and hungry = 0 and status='1'")
        data = cursor.fetchone(); 
        if data is not None:
            dbcnn.fun_db_open();
            SQL = "UPDATE ap_userpet SET status = '2' WHERE Del_YN='N' and hungry = 0 and status='1'"
            cursor.execute(SQL);db.commit();dbcnn.fun_db_close();print("寵物已逃走")
        #--再次執行timer_hg裡的fun_hungry事件
        timer_hg.start();

#--將class實體化
timer1=timer()
#--執行第一次的fun_hungry事件
timer1.fun_hungry()
#--執行第一次的fun_push事件
timer1.fun_push()
```
